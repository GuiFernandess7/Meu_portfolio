## Utilizando DataFrames Externos

```python
csv_file_list = []

def get_df(file_path, sep):
    df = pd.read_csv(file_path, sep=sep)
    return df

def csv_files(path):
    for csv_file in os.listdir(path):
        yield csv_file
        
for file in csv_files('datasets'):
    csv_file_list.append(file)
    
life_expectancy_file = f'datasets\{csv_file_list[1]}'
df = get_df(life_expectancy_file, sep=';')
df
```

```python
min_value =  60.897
max_value = 74.288	
filtered_df = df[(df['Life Expectancy'] >= float(min_value) ) & (df['Life Expectancy'] <= float(max_value))]
filtered_df2 = df.loc[(df["Life Expectancy"] >= min_value)]
filtered_df3 = df.loc[0: min_value]
filtered_df4 = df.filter(df['Life Expectancy'] < min_value, axis=0)
filtered_df4
```
