```python
import pandas as pd
```

```python
df = pd.read_csv("./datasets/World-Population-(2022).csv")
df
```

### Convertendo um NpArray em Lista e extraindo os Itens (str) de uma coluna

```python
df['Country (or dependency)'] = df['Country (or dependency)'].values.astype(str)
countries = df['Country (or dependency)'].values.tolist()
countries_list = [{'Country': country, 'Type': type(country)} for country in countries]
countries_list
```

### Extraindo os Itens (str) de várias colunas
#### Note que dessa vez os dados da coluna "População" são tradados: convertidos para tipo inteiro

```python
df["Population (2022)"] = df.loc[:, "Population (2022)"].replace(",", "").apply(pd.to_numeric, errors='coerce')
countries_and_populations = df[["Country (or dependency)", "Population (2022)"]].values.tolist()
countries_and_populations
```

### Usando diferentes tipos de iteração/loops no DataFrame
#### ATENÇÃO : O uso de loops em Dataframes não é recomendado. Mostrarei outras formas de iteração a seguir.

#### Iterar sobre index

```python
# Captura todos os índices
for ind in df.index:
    #print(ind)
    pass

for ind in df.index:
    print(df['Country (or dependency)'][ind], df['Population (2022)'][ind])
```

#### Iterar sobre cada linha com 'range' e 'len'

```python
# Com .loc[] -- Seleciona a coluna por nome 
for i in range(len(df)):
    #print(df.loc[i, "Country (or dependency)"], df.loc[i, "Population (2022)"])
    pass

# Com .iloc[] -- Seleciona a coluna por index numérico
for i in range(len(df)):
    print(df.iloc[i, 0], df.iloc[i, 2])
```

#### Iterar sobre cada linha com Iterrows()

```python
for index, row in df.iterrows():
    print(row["Country (or dependency)"], row["Population (2022)"])
```

#### Iterar sobre cada linha com Itertuples()

```python
for row in df.itertuples():
    print(row)
```

#### Iterar sobre cada linha com Apply() -- Melhor forma

```python
app_df = df.apply(lambda row: row["Population (2022)"], axis=1)
app_df
```
